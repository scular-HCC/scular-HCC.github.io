## Key Points:
- The generic complex number $z=r(\cos\theta + j\sin\theta)$ can be also written as $z=re^{j\theta}$.
- As is the case with real exponentials, $e^{j(\theta+\lambda)}=e^{j\theta}e^{j\lambda}$.
- The functions $\cos\theta$ and $\sin\theta$ can be expressed in terms of complex exponentials:

$$
\cos\theta = \frac{e^{j\theta}+e^{-j\theta}}{2}
$$

$$
\sin\theta = \frac{e^{j\theta}-e^{-j\theta}}{2j}
$$

- The equation $z^n=e^{j\lambda}$, where $\lambda$ is a given angle, has $n$ roots of the form $z=e^{j\theta}$; these are obtained by setting

$$
\theta = \frac{\lambda + 2k\pi}{n}, \quad k=0,\ldots,n-1.
$$

- The functions $\cos\theta$ and $\sin\theta$ are both periodic with period $2\pi$ (radians), and are shifted versions of each other.
- The generic time-dependent sinusoid $A\cos(\omega t+\phi)$ has three parameters: amplitude $A$, angular frequency $\omega$ (rad/sec) and initial phase $\phi$. The cyclic frequency $f$ (Hz) and period $T$ (sec) are related to $\omega$ by

$$
f = \frac{1}{T} = \frac{\omega}{2\pi}.
$$


## Theory and Examples:


## 1.
Take a complex number $z$ with modulus $|z|=r$ and angle $\angle z=\theta$:

$$
z=r(\cos\theta+j\sin\theta)
$$

An alternative form for $z$ is

$$
z=re^{j\theta}
$$

where we use the identity

$$
e^{j\theta}=\cos\theta+j\sin\theta.
$$

This can be obtained from the Taylor series

$$
e^t = 1 + t + \frac{t^2}{2!} + \frac{t^3}{3!} + \cdots
$$

by setting $t=j\theta$. Grouping the real and imaginary terms separately on the expansion side, we recognize the two Taylor series

$$
\cos\theta = 1-\frac{\theta^2}{2!}+\frac{\theta^4}{4!}-\cdots \quad (	ext{real part})
$$

$$
\sin\theta = \theta-\frac{\theta^3}{3!}+\frac{\theta^5}{5!}-\cdots \quad (	ext{imaginary part})
$$


## 2. Example.
In polar form, $z_1=-1+j$ and $z_2=1+j\sqrt{3}$ are given by

$$
|z_1|=\sqrt{2},\quad \angle z_1=\frac{3\pi}{4}
$$

and

$$
|z_2|=2,\quad \angle z_2=\frac{\pi}{3}.
$$

In complex exponential form,

$$
z_1=\sqrt{2}\,e^{j(3\pi/4)}\quad	ext{and}\quad z_2=2\,e^{j(\pi/3)}.
$$


```python
import numpy as np

# Verify the values in Example 2
z1 = -1 + 1j
z2 = 1 + 1j*np.sqrt(3)
abs_z1, ang_z1 = np.abs(z1), np.angle(z1)
abs_z2, ang_z2 = np.abs(z2), np.angle(z2)
abs_z1, ang_z1, abs_z2, ang_z2
```

$$

## 5.
The equation $z^n=v$, where $z$ is a complex variable and $v$ is a complex constant, has $n$ complex roots (this is true for any polynomial of degree $n$).

These roots can be found using the following observations:

- $|z|^n=|v|$, therefore every root $z$ must satisfy $|z|=|v|^{1/n}$, i.e., it must lie on a circle of radius $|v|^{1/n}$ centered at the origin;
- on that circle, the complex number $z$ with $\angle z=(\angle v)/n$ is a root;
- on the same circle, the angles $(\angle v)/n + k(2\pi/n)$, $k=1,\ldots,n-1$ also correspond to roots (the remaining $n-1$ roots of the equation).


## 6. Example.
To determine the four roots of

$$
z^4=j,
$$

we note that $j=e^{j\pi/2}$. Since $|j|=1$, all the roots of the equation lie on the unit circle. And since $\angle j=\pi/2$, the roots are

$$
e^{j(\pi/8)},\; e^{j(5\pi/8)},\; e^{j(9\pi/8)}\;	ext{and}\; e^{j(13\pi/8)}.
$$

**Your task:** Determine and sketch the roots of

$$
z^3=-8.
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Roots of z^4 = j
angles4 = np.array([np.pi/8, 5*np.pi/8, 9*np.pi/8, 13*np.pi/8])
roots4 = np.exp(1j*angles4)

# Roots of z^3 = -8
angles3 = (np.pi + 2*np.pi*np.array([0,1,2]))/3
roots3 = 2*np.exp(1j*angles3)

th = np.linspace(0, 2*np.pi, 600)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# z^4=j
ax[0].plot(np.cos(th), np.sin(th))
ax[0].scatter(np.real(roots4), np.imag(roots4))
ax[0].axhline(0); ax[0].axvline(0)
ax[0].set_aspect('equal', 'box')
ax[0].grid(True)
ax[0].set_title('Roots of $z^4=j$')

# z^3=-8
ax[1].plot(2*np.cos(th), 2*np.sin(th))
ax[1].scatter(np.real(roots3), np.imag(roots3))
ax[1].axhline(0); ax[1].axvline(0)
ax[1].set_aspect('equal', 'box')
ax[1].grid(True)
ax[1].set_title('Roots of $z^3=-8$')

plt.show()
roots3
```

## 7.
The functions $\cos\theta$ and $\sin\theta$ (where $\theta$ is in radians) are both periodic with period $2\pi$:

$$
\cos(\theta+2\pi)=\cos\theta
$$

$$
\sin(\theta+2\pi)=\sin\theta.
$$

The cosine has (even) symmetry about $\theta=0$, the sine has odd symmetry (or antisymmetry):

$$
\cos(-\theta)=\cos\theta
$$

$$
\sin(-\theta)=-\sin\theta.
$$

Either function can be obtained from the other by shifting $\theta$ by $\pi/2$ in the appropriate direction:

$$
\sin\theta=\cos\left(\theta-\frac{\pi}{2}\right)
$$

$$
\cos\theta=\sin\left(\theta+\frac{\pi}{2}\right).
$$

A shift in $\theta$ by $\pi$ (same as $-\pi$) results in sign reversal in each case:

$$
\cos(\theta+\pi)=-\cos\theta
$$

$$
\sin(\theta+\pi)=-\sin\theta.
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Editable plot corresponding to the lecture figure
theta = np.linspace(0, 4*np.pi, 1000)
fig, ax = plt.subplots(1,2, figsize=(10,3))
ax[0].plot(theta, np.cos(theta))
ax[0].set_title('cosθ')
ax[0].set_xlabel('θ')
ax[0].grid(True)
ax[1].plot(theta, np.sin(theta))
ax[1].set_title('sinθ')
ax[1].set_xlabel('θ')
ax[1].grid(True)
plt.show()
```

## 8.
The generic continuous-time sinusoid is given by

$$
x(t)=A\cos(\omega t+\phi)
$$

where
- $A>0$ is the amplitude
- $\omega$ is the angular frequency, in rad/sec
- $t$ is in seconds
- $\phi$ is the initial phase, in radians

Since $\cos\theta$ has period $2\pi$ radians, $x(t)$ has period

$$
T=\frac{2\pi}{\omega}\ 	ext{seconds}.
$$

The cyclic frequency (i.e., periods or cycles per second) is given by

$$
f=\frac{1}{T}=\frac{\omega}{2\pi}
$$

and is expressed in Hz.


## 9. Example.
Suppose that you are given the following information about a continuous-time sinusoid $x(t)=A\cos(\omega t+\phi)$:

- $x(t)\ge 1.50$ for $23.0\%$ of its period;
- it takes $0.02$ seconds for the value of the sinusoid to drop from $1.50$ to $0.00$; and
- $x(0)=-0.3$ and the first derivative $x'(0)$ is positive.

You have three (and then some) independent pieces of information and three unknowns $(A,\omega,\phi)$, so you should be able to solve for these parameters.

Justify each of the following results:

- The first piece of information gives

$$
A=\frac{1.5}{\cos(0.23\pi)}=2.0.
$$

- In addition, the second piece of information gives

$$
\frac{0.5\pi-0.23\pi}{\omega}=0.02\ \Rightarrow\ \omega=13.5\pi.
$$

**Question:** What are the values of $f$ (in Hz) and $T$ (in sec)?

- With $A$ known, the third piece of information gives

$$
\phi=-\cos^{-1}\left(\frac{-0.3}{2.0}\right)=-1.721\ 	ext{rad}.
$$


```python
import numpy as np

A = 1.5/np.cos(0.23*np.pi)
omega = (0.5*np.pi - 0.23*np.pi)/0.02
f = omega/(2*np.pi)
T = 1/f
phi = -np.arccos((-0.3)/A)
A, omega, f, T, phi
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Editable plot of the sinusoid over one period using the computed parameters
A = float(A)
omega = float(omega)
phi = float(phi)
T = 2*np.pi/omega

t = np.linspace(0, T, 1000)
x = A*np.cos(omega*t + phi)

plt.figure(figsize=(8,3))
plt.plot(t, x)
plt.axhline(0)
plt.grid(True)
plt.title('x(t) over one period')
plt.xlabel('t (sec)')
plt.ylabel('x(t)')
plt.show()
```
