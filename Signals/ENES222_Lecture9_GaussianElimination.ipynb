```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# -----------------------------
# Lecture 9: Gaussian Elimination
# -----------------------------

# Rows of augmented matrix [A | b]
R1 = np.array([2, 1, 2])   # [2  1 | 2]
R2 = np.array([4, 3, 6])   # [4  3 | 6]

# Target row after elimination
R2_final = R2 - 2 * R1

# Setup plot
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 7)
ax.set_aspect('equal')
ax.grid(True)
ax.set_title("Lecture 9: Gaussian Elimination (Row Operation)")

# Plot vectors
row1, = ax.plot([], [], lw=3, color="blue", label="Row 1")
row2, = ax.plot([], [], lw=3, color="red", label="Row 2")

ax.legend(loc="upper left")

def update(frame):
    # Row 1 is constant
    row1.set_data([0, R1[0]], [0, R1[2]])

    # Animate row 2 transformation
    if frame < 40:
        alpha = frame / 40
        current_row = (1 - alpha) * R2 + alpha * (R2 - 2 * R1)
    else:
        current_row = R2_final

    row2.set_data([0, current_row[0]], [0, current_row[2]])
    return row1, row2

anim = FuncAnimation(fig, update, frames=60, interval=100)
HTML(anim.to_jshtml())
```

This animation shows Gaussian elimination by visually implementing the row operation
R2←R2−αR1R_2 \leftarrow R_2 - \alpha R_1R2​←R2​−αR1​ as a vector subtraction, making the algebraic elimination step geometrically explicit.
