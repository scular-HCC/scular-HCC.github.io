## Key Points
- Two complex numbers can be multiplied by expressing each number in the form

$$
z = x + jy
$$

then using distributivity and the rule

$$
j^2 = -1.
$$

- The product of two complex numbers with polar coordinates $(r_1,\theta_1)$ and $(r_2,\theta_2)$ is the complex number with polar coordinates

$$
(r_1 r_2,\; \theta_1+\theta_2).
$$

- The product of a complex number $z$ and its conjugate $z^*$ equals the square modulus

$$
|z|^2.
$$

- If $z$ has polar coordinates $(r,\theta)$, its inverse $z^{-1}$ has polar coordinates

$$
(r^{-1},\; -\theta).
$$


## 1.
The modulus $|z|$ represents the length of the vector $z$. Similarly, the expression

$$
|z_1 - z_2|
$$

gives the distance between the points $z_1$ and $z_2$.

Using this concept, we see that the following two equations in the variable $z$ each have a familiar geometrical interpretation:

- 
$$
|z - z_0| = a,
$$

where $a$ is a positive constant, is the set of points $z$ on the complex plane which are at a fixed distance $a$ from the point $z_0$. This is a circle of radius $a$ centered at $z_0$.

- 
$$
|z - z_1| = |z - z_2|
$$

is the set of points $z$ on the complex plane which are equidistant from the points $z_1$ and $z_2$. This is the perpendicular bisector of the line segment joining $z_1$ and $z_2$.


**Your task:** Sketch the two curves given by the equations

$$
|z-(3+4j)|=5
$$

and

$$
|z-3|=|z+4j|.
$$


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-6, 10, 500)
y = np.linspace(-6, 10, 500)
X, Y = np.meshgrid(x, y)

circle = (X-3)**2 + (Y-4)**2
bisector = np.sqrt((X-3)**2 + Y**2) - np.sqrt(X**2 + (Y+4)**2)

plt.figure(figsize=(6,6))
plt.contour(X,Y,circle,levels=[25],colors='blue')
plt.contour(X,Y,bisector,levels=[0],colors='red')
plt.axhline(0)
plt.axvline(0)
plt.axis('equal')
plt.grid(True)
plt.title('Lecture 2: Geometry on the Complex Plane')
plt.show()
```

## 2.
The most common form for a complex number $z$ incorporates the real and imaginary parts as follows:

$$
z = x + jy.
$$

This form, together with the convention that

$$
j^2 = -1,
$$

allows us to multiply two complex numbers together. For example,

$$
(5-2j)(3-4j)=15-20j-6j+8j^2=7-26j.
$$


## 3.
In polar form, multiplication of complex numbers is simple. If

$$
z_1 = r_1(\cos\theta_1 + j\sin\theta_1)
$$

and

$$
z_2 = r_2(\cos\theta_2 + j\sin\theta_2),
$$

then

$$
z_1 z_2 = r_1 r_2\left(\cos\theta_1\cos\theta_2-\sin\theta_1\sin\theta_2)
+ j(\cos\theta_1\sin\theta_2+\sin\theta_1\cos\theta_2)\right).
$$

Using the identities

$$
\cos\theta_1\cos\theta_2-\sin\theta_1\sin\theta_2=\cos(\theta_1+\theta_2)
$$

and

$$
\cos\theta_1\sin\theta_2+\sin\theta_1\cos\theta_2=\sin(\theta_1+\theta_2),
$$

we see that $z_1z_2$ has modulus $r_1r_2$ and angle $\theta_1+\theta_2$.


## 4. Example.
As before, take

$$
z_1=5-2j, \quad z_2=3-4j.

The polar forms are

$$

and, in Cartesian form,

$$
z_1z_2=5\sqrt{29}\left(\cos(-1.3078)+j\sin(-1.3078)\right)=7-26j.

## 5.
Division of two complex numbers makes implicit use of the inverse, i.e.,

$$
\frac{z_1}{z_2}=z_1 z_2^{-1}.
$$

This can be carried out in a single step by multiplying both the numerator $z_1$ and the denominator $z_2$ by the complex conjugate of $z_2$.

If

$$
z_1=5-2j, \quad z_2=3-4j,
$$

then

$$
\frac{z_1}{z_2}= \frac{(5-2j)(3+4j)}{(3-4j)(3+4j)}
= \frac{(5-2j)(3+4j)}{(3^2)(4^2)}
=\frac{23}{25}+j\frac{14}{25}.
$$

Here we used the identity

$$
z z^* = |z|^2.
$$

i.e., the product of a complex number and its conjugate equals the (real-valued) square magnitude of that number.

## 6.
In polar form, dividing two complex numbers corresponds to dividing their magnitudes and subtracting one angle from the other. This is because

$$
z^{-1}=\frac{z^*}{|z|^2},
$$

and therefore

$$
|z^{-1}|=|z|^{-1}, \quad \angle z^{-1}=-\angle z.
$$

**Your task:** Verify the above results. Plot $z$, $z^*$, and $z^{-1}$ for

$$
z=3-4j.
$$


```python
z = np.array([3,-4])
z_conj = np.array([3,4])
z_inv = z_conj/(3**2+4**2)

plt.figure(figsize=(6,6))
plt.axhline(0)
plt.axvline(0)
plt.plot(z[0],z[1],'ro',label='z')
plt.plot(z_conj[0],z_conj[1],'bo',label='z*')
plt.plot(z_inv[0],z_inv[1],'go',label='z^{-1}')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.title('z, z* and z^{-1}')
plt.show()
```
