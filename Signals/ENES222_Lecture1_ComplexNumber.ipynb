```python
from matplotlib import rc
rc('animation', html='jshtml')

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Complex number from Lecture 1
x = -5
y = 2
z = x + 1j*y

# Magnitude and angle
r = np.sqrt(x**2 + y**2)
theta = np.arctan2(y, x)

# Set up the figure
fig, ax = plt.subplots(figsize=(6,6))
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_xlabel("Real axis")
ax.set_ylabel("Imaginary axis")
ax.set_title("Animation of Complex Number z = -5 + 2j")

# Draw axes
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Elements to animate
point, = ax.plot([], [], 'ro', markersize=8)
vector, = ax.plot([], [], 'r-', linewidth=2)
angle_arc, = ax.plot([], [], 'b-', linewidth=2)

# Animation function
def update(frame):
    t = frame / 100  # progress from 0 to 1
    x_t = t * x
    y_t = t * y
    
    # Update point and vector
    point.set_data([x_t], [y_t])
    vector.set_data([0, x_t], [0, y_t])
    
    # Angle arc
    arc_theta = np.linspace(0, t * theta, 100)
    arc_x = 1.5 * np.cos(arc_theta)
    arc_y = 1.5 * np.sin(arc_theta)
    angle_arc.set_data(arc_x, arc_y)
    
    return point, vector, angle_arc

ani = FuncAnimation(fig, update, frames=100, interval=40, blit=True)

ani
```
