```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# -----------------------------
# Lecture 10: Orthogonal Projection
# (same animation style as Lecture 9)
# -----------------------------

# Vectors
a = np.array([1.0, 0.0])
b = np.array([1.0, 1.0])

# Projection
proj_b = (np.dot(a, b) / np.dot(a, a)) * a

# Plot setup
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-0.2, 1.4)
ax.set_ylim(-0.2, 1.4)
ax.set_aspect('equal')
ax.grid(True)
ax.set_title("Lecture 10: Orthogonal Projection")

# Static vectors
va, = ax.plot([], [], lw=3, color="blue", label="a")
vb, = ax.plot([], [], lw=3, color="black", label="b")

# Animated vectors
vproj, = ax.plot([], [], lw=3, color="green", label="projₐ(b)")
verr,  = ax.plot([], [], lw=2, color="red", linestyle="--", label="b − projₐ(b)")

ax.legend(loc="upper left")

# -----------------------------
# Animation update (Lecture 9 style)
# -----------------------------
def update(frame):
    # Always draw original vectors
    va.set_data([0, a[0]], [0, a[1]])
    vb.set_data([0, b[0]], [0, b[1]])

    # Animate projection
    alpha = min(frame / 40, 1.0)
    p = alpha * proj_b

    vproj.set_data([0, p[0]], [0, p[1]])
    verr.set_data([p[0], b[0]], [p[1], b[1]])

    return va, vb, vproj, verr

anim = FuncAnimation(fig, update, frames=60, interval=100)

HTML(anim.to_jshtml())
```
