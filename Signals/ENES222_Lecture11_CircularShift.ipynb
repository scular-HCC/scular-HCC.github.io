```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# -----------------------------
# Lecture 11: Circular Shift Animation
# (same animation style as Lecture 9)
# -----------------------------

# Original signal
x = np.array([1, 2, 3, 2, 1])
N = len(x)

# Circular shift index
shifts = np.arange(N)

# Setup plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.set_xlim(-0.5, N - 0.5)
ax.set_ylim(0, 3.5)
ax.grid(True)
ax.set_title("Lecture 11: Circular Shift via Permutation Matrix")

# Plot initial signal
line, = ax.plot(np.arange(N), x, 'o-', lw=3, label="x[n]")

ax.legend()

# -----------------------------
# Animation update (Lecture 9 style)
# -----------------------------
def update(frame):
    # Circular shift
    x_shift = np.roll(x, frame)

    # Update plot
    line.set_data(np.arange(N), x_shift)

    return (line,)

anim = FuncAnimation(fig, update, frames=N, interval=500)

HTML(anim.to_jshtml())
```
