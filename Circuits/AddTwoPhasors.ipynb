```python

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import Play, IntSlider, VBox, Output, jslink
from IPython.display import display, clear_output
```

```python
# Phasor magnitudes
A_mag = 1.0
B_mag = 0.7

# Initial phase offsets (radians)
phi_A = 0
phi_B = np.pi / 4

# Angular speed
omega = 2 * np.pi      # rad/s

# Animation setup
frames = 200
```

```python
out = Output()
display(out)
```

```python
def draw_frame(frame):
    t = frame / frames
    theta = omega * t
    
    # Complex phasors
    A = A_mag * np.exp(1j * (theta + phi_A))
    B = B_mag * np.exp(1j * (theta + phi_B))
    R = A + B
    
    # Magnitudes and angles
    A_mag_val, A_ang = np.abs(A), np.degrees(np.angle(A))
    B_mag_val, B_ang = np.abs(B), np.degrees(np.angle(B))
    R_mag_val, R_ang = np.abs(R), np.degrees(np.angle(R))
    
    with out:
        clear_output(wait=True)
        
        fig, ax = plt.subplots(figsize=(6, 6))
        
        ax.axhline(0, color='gray', linewidth=0.5)
        ax.axvline(0, color='gray', linewidth=0.5)
        
        # Draw phasors
        ax.arrow(0, 0, A.real, A.imag,
                 head_width=0.06, length_includes_head=True,
                 color='blue')
        
        ax.arrow(A.real, A.imag, B.real, B.imag,
                 head_width=0.06, length_includes_head=True,
                 color='green')
        
        ax.arrow(0, 0, R.real, R.imag,
                 head_width=0.06, length_includes_head=True,
                 color='red')
        
        # Text labels
        ax.text(A.real/2, A.imag/2,
                f"A = {A_mag_val:.2f} ∠ {A_ang:.1f}°",
                color='blue', fontsize=10)
        
        ax.text(A.real + B.real/2, A.imag + B.imag/2,
                f"B = {B_mag_val:.2f} ∠ {B_ang:.1f}°",
                color='green', fontsize=10)
        
        ax.text(R.real/2, R.imag/2,
                f"R = {R_mag_val:.2f} ∠ {R_ang:.1f}°",
                color='red', fontsize=10, fontweight='bold')
        
        ax.set_aspect('equal', adjustable='box')
        ax.set_xlim(-2, 2)
        ax.set_ylim(-2, 2)
        ax.set_title("Adding Two Phasors\nwith Magnitude and Angle")
        ax.set_xlabel("Real Axis")
        ax.set_ylabel("Imaginary Axis")
        ax.grid(True)
        
        plt.show()
```

```python
play = Play(
    value=0,
    min=0,
    max=frames - 1,
    step=1,
    interval=80
)

slider = IntSlider(min=0, max=frames - 1)
jslink((play, "value"), (slider, "value"))

slider.observe(lambda c: draw_frame(c["new"]), names="value")

display(VBox([play]))
draw_frame(0)
```
