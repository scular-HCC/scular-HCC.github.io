```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
```

```python
def capacitor_phase_dual_axis(C=1e-6):
    V0 = 1.0
    f = 1.0
    omega = 2 * np.pi * f
    
    t = np.linspace(0, 2, 500)
    
    v = V0 * np.sin(omega * t)
    i = omega * C * V0 * np.cos(omega * t)
    
    fig, ax1 = plt.subplots(figsize=(10, 5))
    
    ax1.plot(t, v, 'b', label="Voltage (V)", linewidth=2)
    ax1.set_xlabel("Time (s)")
    ax1.set_ylabel("Voltage (V)", color='b')
    ax1.tick_params(axis='y', labelcolor='b')
    
    ax2 = ax1.twinx()
    ax2.plot(t, i, 'r', label="Current (A)", linewidth=2)
    ax2.set_ylabel("Current (A)", color='r')
    ax2.tick_params(axis='y', labelcolor='r')
    
    fig.suptitle("Capacitor Phase Relationship\n(Current Leads Voltage by 90°)")
    ax1.grid(True)
    
    fig.tight_layout()
    plt.show()

interact(
    capacitor_phase_dual_axis,
    C=FloatSlider(
        value=1e-6,
        min=0.1e-6,
        max=5e-6,
        step=0.1e-6,
        description="Capacitance (F)",
        readout_format=".1e"
    )
);
```
