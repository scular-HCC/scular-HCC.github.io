```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
L = 0.1          # Inductance (H)
I0 = 1.0         # Current amplitude (A)
f = 1.0          # Frequency (Hz)
omega = 2 * np.pi * f

# Time
t = np.linspace(0, 2, 500)

# Signals
current = I0 * np.sin(omega * t)
voltage = omega * L * I0 * np.cos(omega * t)  # v = L di/dt

# Plot
fig, ax1 = plt.subplots(figsize=(10, 5))

# Voltage axis
ax1.plot(t, voltage, 'b', linewidth=2, label="Voltage (V)")
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("Voltage (V)", color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Current axis
ax2 = ax1.twinx()
ax2.plot(t, current, 'r', linewidth=2, label="Current (A)")
ax2.set_ylabel("Current (A)", color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Title & grid
fig.suptitle("Inductor Phase Relationship\nVoltage Leads Current by 90°")
ax1.grid(True)

# Combined legend
lines = ax1.get_lines() + ax2.get_lines()
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc="upper right")

plt.tight_layout()
plt.show()
```
